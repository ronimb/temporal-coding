```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns; sns.set()
import numpy as np
```

Specify data location and load the data

```python
data_location = '/home/ron/OneDrive/Documents/Masters/Parnas/temporal-coding/Results/exp_results.pkl'
data = pd.read_pickle(data_location)
data.head()
```

```python
sns.relplot(x='orig_difference', y='post', data=data, col='frequency',
            style='release_duration', hue='release_duration', legend='full',
            kind='line', height=5, aspect=1, );
```

```python
sns.relplot(x='release_duration', y='post', data=data, col='frequency',
            style='orig_difference', hue='orig_difference', 
            kind='line', height=5, aspect=1, 
            hue_order=['(1, 3)', '(3, 5)', '(5, 7)', '(7, 9)']);
```

```python
ax = sns.catplot(x='orig_difference', y='post', hue='release_duration', data=data, height=6, aspect=4, kind='boxen',
                order=['(1, 3)', '(3, 5)', '(5, 7)', '(7, 9)'], row='frequency');
```
